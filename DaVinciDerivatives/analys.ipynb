```python
def findTtlProfit(c, M, stockA, stockB):
    idxA = 0
    idxB = 0
    profit = 0
    while idxA < len(stockA) and idxB < len(stockB):
        # if idxA/10 == idxA//10:
        #     print(f'A: {idxA}', end=",")
        # if idxB/10 == idxB//10: 
        #     print(f'B: {idxB}', end=",")
        currTimeA = stockA.iloc[idxA, 0]
        currTimeB = stockB.iloc[idxB, 0]
        # print(idxA, idxB)
        if currTimeA < currTimeB and idxA < len(stockA) -  1:
            idxA += 1
            print('statement check A', stockA.iloc[idxA, 0] < currTimeB)
            if (stockA.iloc[idxA, 1] - stockA.iloc[(idxA - 1), 1]) >= c:
                
                buyTime = stockA.iloc[idxA, 0] + timedelta(milliseconds=M)
                print('hitA', idxA, buyTime, 'when time B is', idxB, currTimeB)
                #print('statement check', buyTime < currTimeB)
                profit += make_transaction(stockB, idxB, buyTime)

        elif currTimeA > currTimeB and idxB < len(stockB) - 1:
            idxB += 1
            print('statement check B', stockB.iloc[idxB, 0] < currTimeA)
            if (stockB.iloc[idxB, 1] - stockB.iloc[(idxB - 1), 1]) >= c:
                buyTime = stockB.iloc[idxB, 0] + timedelta(milliseconds=M)
                print('hitB', idxB, buyTime, 'when time A', idxA, currTimeA)
                #print('statement check', buyTime < currTimeA)
                profit += make_transaction(stockA, idxA, buyTime)
        else:
            idxA += 1
            idxB += 1
    return profit

def make_transaction(stock, currTime, buyTime):        
    if buyTime > stock.iloc[currTime, 0]:
        # check inbound
        if currTime > len(stock): 
            return 0
        gain = stock.iloc[currTime + 1, 1] - stock.iloc[currTime, 1]
    elif buyTime < stock.iloc[currTime, 0]:
        gain = stock.iloc[currTime, 1] - stock.iloc[currTime - 1, 1]
    return gain
```
